## I. Prerequisites

- ###  Java: Apache Kafka is written in Java, so you need to have Java Development Kit (JDK) installed on your Windows machine. You can download and install the latest version of the JDK from the Oracle website or adopt OpenJDK.

- ###  ZooKeeper: Apache Kafka relies on Apache ZooKeeper for distributed coordination. While ZooKeeper is not mandatory for running a single Kafka broker (for development purposes), it’s recommended to have ZooKeeper installed. You can download it from the Apache ZooKeeper website.

## II. Steps:

- ### Step 1: Install Kafka binary source file in link: https://kafka.apache.org/downloads



<br></br>
- ### Step 2: Extract file to C:\kafka (for example: C:\Kafka\kafka_2.13-3.7.0)
<br></br>
- ### Step 3: Update the Environment variable
    - ####  3.1 Press Windows key and search "edit the system environment variables"
    - ####  3.2 Environment.. -> In system variables section find the "path" and click Edit
     #### 
    - ####  3.3 Copy the path: C:\Kafka\kafka_2.13-3.7.0\bin\windows then click New -> paste the path
     #### 
    - ####  3.4 Click Ok -> Ok -> Ok
<br></br>
- ### Step 4: Edit Config files
    - #### 4.1 Open path: C:\Kafka\kafka_2.13-3.7.0\config and find Server.properties and zookeeper.properties
    
    - #### 4.2 Edit the Server.properties like below correspond to your path
    

    - #### 4.2 Edit the zookeeper.properties like below correspond to your path
    
<br></br>
- ### Step 5: Start the Zookeeper
    - #### 5.1: Open a command prompt and navigate to your ZooKeeper installation directory (C:\Kafka\kafka_2.13-3.7.0\bin\windows)
    - #### 5.2: Type ```bash zookeeper-server-start ..\..\config\zookeeper.properties ``` and Enter
<br></br>
- ### Step 6: Start the Kafka Broker
    - #### 6.1: Open a command prompt and navigate to your kafka-server-start.bat installation directory (C:\Kafka\kafka_2.13-3.7.0\bin\windows)
    - #### 6.2: Type: ```bash kafka-server-start ..\..\config\server.properties ``` and Enter
<br></br>
- ### Step 7: Create a Kafka Topic
    - #### 7.1: Open another command prompt and type
    - #### ```bash kafka-topics --bootstrap-server localhost:9092 --topic first_topic --create --partitions 3 --replication-factor 1 ``` -> Enter
<br></br>
- ### Step 8: Produce Messages
    - #### 8.1: Open a command prompt and type
    - #### ```bash kafka-console-producer --bootstrap-server localhost:9092 --topic first_topic ``` -> Enter
<br></br>
- ### Step 9: Consumer Messages
    - #### 9.1: Open a command prompt and type
    - #### ```bash kafka-console-consumer --bootstrap-server localhost:9092 --topic first_topic ``` -> Enter
<br></br>
- ### Step 10: Testing Streaming Data
    - #### 10.1: Type the message in the Produce Messages command prompt open in step 8
    - #### You will the message in the Consumer command prompt that we open in step 9
<br></br>
<br></br>

- ### Note:
### To List all topic, open command prompt type ```bash kafka-topics — bootstrap-server localhost:9092 — list```
