```python
def get_return(coin):
    coin_frame = df[df.coin == coin]
    coin_frame = coin_frame.sort_values('timestamp_converted')
    coin_frame  = coin_frame.set_index('timestamp_converted')
    return coin_frame
```

```python
rets = []

df =  pd.read_csv('CoinPrices.csv',header=None)
df.columns = ['timestamp_converted', 'price', 'coin']

for coin in df.coin.unique():
    rets.append(get_return(coin))

series_ = pd.Series(rets, index=df.coin.unique())
```
